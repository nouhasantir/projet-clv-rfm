```python
import pandas as pd

def compute_kpis_global(cohort_counts: pd.DataFrame, cohort_revenue: pd.DataFrame):
  
    acq_col = "AcqMonth"
    
    # Colonnes d'âge
    age_cols_counts = [c for c in cohort_counts.columns if c != acq_col]
    age_cols_revenue = [c for c in cohort_revenue.columns if c != acq_col]
    
    # Nombre total de clients acquis = somme des counts à l'âge 0
    if "0" in cohort_counts.columns:
        nb_clients_acquis = cohort_counts["0"].sum()
    else:
        raise KeyError("La colonne '0' (âge 0) est absente de cohort_counts.")
    
    # CA total sur toutes les cohortes et tous les âges
    ca_total = cohort_revenue[age_cols_revenue].sum().sum()
    
    # CLV moyenne approximative (CA total / nb de clients acquis)
    clv_moyenne = ca_total / nb_clients_acquis if nb_clients_acquis > 0 else None
    
    return {
        "clients_acquis": nb_clients_acquis,
        "ca_total": ca_total,
        "clv_moyenne": clv_moyenne,
    }


def get_retention_matrix(cohort_counts: pd.DataFrame):
 
    acq_col = "AcqMonth"
    if acq_col not in cohort_counts.columns:
        raise KeyError("La colonne 'AcqMonth' est absente de cohort_counts.")
    
    # Colonnes d'âge (toutes sauf AcqMonth)
    age_cols = [c for c in cohort_counts.columns if c != acq_col]
    
    if "0" not in cohort_counts.columns:
        raise KeyError("La colonne '0' (âge 0) est absente de cohort_counts.")
    
    # Matrice de rétention : chaque ligne divisée par sa valeur à l'âge 0
    retention = cohort_counts[age_cols].div(cohort_counts["0"], axis=0)
    
    # On pose AcqMonth en index pour que ce soit prêt pour une heatmap
    retention.insert(0, acq_col, cohort_counts[acq_col])
    retention = retention.set_index(acq_col)
    
    return retention
```

Interprétation des KPIs globaux

Les indicateurs affichés ci-dessus permettent d’obtenir une vue d’ensemble de la performance business à partir des données agrégées par cohortes.

1. Nombre total de clients acquis

Ce KPI correspond à la somme des effectifs à l’âge 0 (colonnes “0” de cohort_counts).
Cela représente l’ensemble des nouveaux clients qui ont effectué au moins un achat durant la période analysée.
C’est la base populationnelle utilisée pour tous les calculs de rétention et de CLV.

2. Chiffre d’affaires total

Le CA total est obtenu en sommant toutes les valeurs du tableau cohort_revenue (toutes cohortes × tous âges).
Cela donne une estimation du revenu global généré par l’ensemble des clients au fil du temps.

3. CLV moyenne (approximative)

```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path(r"C:\Users\bessy\Documents\projet_clv_rfm\data")

def load_cohort_counts():
    return pd.read_csv(DATA_DIR / "cohort_counts.csv")

def load_cohort_revenue():
    return pd.read_csv(DATA_DIR / "cohort_revenue.csv")  # adapte si c'est revenues

def load_all():
    cohort_counts = load_cohort_counts()
    cohort_revenue = load_cohort_revenue()
    return cohort_counts, cohort_revenue


def compute_kpis_global(cohort_counts: pd.DataFrame, cohort_revenue: pd.DataFrame):
    acq_col = "AcqMonth"
    
    age_cols_counts = [c for c in cohort_counts.columns if c != acq_col]
    age_cols_revenue = [c for c in cohort_revenue.columns if c != acq_col]
    
    if "0" not in cohort_counts.columns:
        raise KeyError("La colonne '0' (âge 0) est absente de cohort_counts.")
    
    nb_clients_acquis = cohort_counts["0"].sum()
    ca_total = cohort_revenue[age_cols_revenue].sum().sum()
    clv_moyenne = ca_total / nb_clients_acquis if nb_clients_acquis > 0 else None
    
    return {
        "clients_acquis": nb_clients_acquis,
        "ca_total": ca_total,
        "clv_moyenne": clv_moyenne,
    }


def get_retention_matrix(cohort_counts: pd.DataFrame):
    acq_col = "AcqMonth"
    if acq_col not in cohort_counts.columns:
        raise KeyError("La colonne 'AcqMonth' est absente de cohort_counts.")
    
    age_cols = [c for c in cohort_counts.columns if c != acq_col]
    
    if "0" not in cohort_counts.columns:
        raise KeyError("La colonne '0' (âge 0) est absente de cohort_counts.")
    
    retention = cohort_counts[age_cols].div(cohort_counts["0"], axis=0)
    retention.insert(0, acq_col, cohort_counts[acq_col])
    retention = retention.set_index(acq_col)
    
    return retention
```

## 🔥 Analyse de la heatmap de rétention

La heatmap représente la **rétention par cohorte d’acquisition**.

- Chaque **ligne** correspond à une cohorte (`AcqMonth`), c’est-à-dire un mois d’acquisition des clients.
- Chaque **colonne** correspond à l’**âge de la cohorte** (0, 1, 2, … mois après l’acquisition).
- Chaque **case** affiche un **taux de rétention** compris entre 0 et 1 :
  
  \[
  \text{Rétention}(cohorte, âge) = \frac{\text{nombre de clients actifs à cet âge}}{\text{nombre de clients à l'âge 0}}
  \]

### Lecture du graphique

- La **colonne 0** est toujours égale à 1 (100 %) par construction : tous les clients sont présents à l’acquisition.
- Plus on avance en âge (M+1, M+2, …), plus les taux diminuent : cela traduit la **perte progressive de clients** (churn).
- Les lignes plus foncées indiquent des cohortes qui **retiennent mieux** leurs clients sur la durée.
- Les différences entre cohortes peuvent révéler :
  - des changements dans la qualité de l’acquisition (source trafic, offre, canal),
  - des effets saisonniers (cohortes de fin d’année vs début d’année),
  - des modifications de politique commerciale ou CRM.

### Intérêt business

Cette visualisation permet à l’équipe marketing de :

- repérer les **âges critiques** où les clients décrochent (forte baisse de rétention),
- identifier les **cohortes performantes** à analyser plus en détail (quelle offre, quel canal ?),
- quantifier l’impact de la rétention sur la **valeur vie client (CLV)**,
- prioriser les actions (relances, campagnes ciblées) autour des moments clés de churn.
